# Creating the graphs and chart from the final dataset

```python
# 1. graph for the count of races of different length
# 2. graph for the count of races for different lengths and gifferent genders
# 3. graph for the distribution of average speed of athlete in the race length of 50 miles
# 4. violin grapg for showing the colume of male and female athlete for their avg speed in different race length
# 5. chart showing the average speed of an thlete in accordance with their age and gender
```

```python
# 1. graph for the count of races of different length
sns. histplot(df_3['race_length'])
```

```python
# 2. graph for the count of races for different lengths and gifferent genders
sns.histplot(df_3, x = 'race_length', hue = 'athlete_gender')
```

```python
# 3. graph for the distribution of average speed of athlete in the race length of 50 miles
sns.displot(df_3[df_3['race_length'] == '50mi']['athlete_average_speed'])
```

```python
# 4. violin grapg for showing the volume of male and female athlete for their avg speed in different race length
sns.violinplot(data = df_3, x = 'race_length', y = 'athlete_average_speed', hue = 'athlete_gender', split = True, inner = 'quartile')
```

```python
# 5. chart showing the average speed of an athlete in accordance with their age and gender
sns.lmplot(data=df_3, x = 'athlete_age', y = 'athlete_average_speed', hue = 'athlete_gender' )
```

#Answering differet questions for the cleaned dataset

```python
df_3.columns
```

```python
# Q. what is the difference in speed for 50km, 50mi male to female?
```

```python
df_3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# Q. what age group are the best in the 50mi race (20+ races min) (show 15)
```

```python
df_3.query('race_length == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False).query('count>19').head(15)
```

```python
# Q. what age group are the worst in the 50mi race (10+ races min) (show 20)
```

```python
df_3.query('race_length == "50mi"').groupby(['athlete_age'])['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = True).query('count>9').head(15)
```

```python
# Seasons for the data -> slower in summer than winter?
# spring 3-5
# summer 6-8
# fall 9-12
# winter 12-2
# split between two decimals
```

```python
df_3['race_month'] = df_3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df_3.head(2)
```

```python
df_3['race_season'] = df_3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df_3.head(2)
```

```python
df_3.groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```

```python
# For 50 miles only
df_3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
